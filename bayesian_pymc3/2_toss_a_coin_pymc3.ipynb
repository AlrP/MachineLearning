```python
import numpy as np
import pymc3 as pm
import arviz as az
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

sns.set()
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
np.random.seed(123)
trials = 4
theta_real = 0.35 # в реальном эксперименте это значение неизвестно
data = stats.bernoulli.rvs(p=theta_real, size=trials)
data
```

Данные сгенерированы. Процесс определения модели включает в себя задание правдободобия и априорной вероятности с помощью распределений вероятностей:

$$Prior: \theta \sim Beta(\alpha, \beta) $$
$$Likelihood: y \sim Bernoulli(n=1, p=\theta)$$

```python
# Context manager creates model container
with pm.Model() as coin_toss_model:
    # Define Beta prior
    theta = pm.Beta('theta', alpha=1, beta=1)
    # Define Bernoulli likelihood
    y = pm.Bernoulli('y', p=theta, observed=data)
    # Make inference sampling 1000 elements from posterior distribution
    trace = pm.sample(1000, random_seed=123)
```

Для выборки используется механизм __NUTS-No U-Turn Sampler__, обрабатывается единственная переменная $\theta$. Но это не самый типичный случай, так как PyMC3 может назначать __различные сэмплеры__ для обработки __различных переменных__.

По 1000 элементов в цепочке предназначено для автоматической точной настройки алгоритма выборки
(в данном примере NUTS). Эта часть выборки отбрасывается по умолчанию (4 цепочки * 1000 элементов). Далее в каждой цепочке обрабатывается 1000 элементов, дающих итоговый результат, таким образом всего генерируется 8000 элементов (2000 - разогрев, 4 цепочки по 1000 - семплинг). Этап точной настройки помогает PyMC3 формировать надежную выборку из апостериорного распределения. Количество этапов точной настройки можно изменить с помощью аргумента tune, передаваемого в функцию sample.

В более старых версиях pymc3 число цепочек и единиц семплинга для разогрева, установленных по-умолчанию,  может отличаться.

#### Trace plots

```python
# Kernel density estimation for each chain on the left
# Sampling process of each chain on the right
az.plot_trace(trace)
```

#### Summary

```python
# Table summary for sampling result
az.summary(trace, hdi_prob=0.95)
```

#### Posterior distribution and credible interval

```python
az.plot_posterior(trace, hdi_prob=0.95)
```

Как видим, количество данных не позволяет сказать что монета является "мошеннической", поскольку значение 0.5 попадает в 95%-ый вероятностный интервал наименьшей длины.

Сравнение 95%-ых вероятностных интервалов наименьшей длины:

HDI 95%, аналитический вывод (1_beta_binomal_model): [0.11, 0.53]\
HDI 95%, NUTS sampler:                               [0.026, 0.685]

#### Region of practical equivalence

Можно ослабить определение подлинности (симметричности) монеты и сказать, что симметричной монете соответствует значение $\theta$, приблизительно равное 0.5. Например, можно утверждать, что любое значение в интервале __[0.45, 0.55]__ будет "практически" равным 0.5 и представлять собой приемлемое решение нашей задачи. 

Такой интервал называют пространством практической равнозначности (ППР) (__Region of Practical Equivalence – ROPE__).

`Пространство практической равнозначности (ППР) – это произвольный интервал, который выбирается на основе второстепенных, менее значимых знаний. При этом предполагается, что любые значения в этом интервале могут считаться практически равнозначными`

Крайний и совсем бессмысленный случай: если задавать ROPE = [0,1], то можно всегда делать вывод о симметричности монеты. Соответственно, само определение ROPE зависит от контекста практической задачи.

#### Posterior distribution and ROPE

```python
az.plot_posterior(trace, hdi_prob=0.95, rope=[0.45, 0.55])
```
