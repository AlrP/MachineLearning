#### Гауссовы статистические выводы

ЯМР позволяет измерять различные типы наблюдаемых количественных явлений, которые связаны с ненаблюдаемыми и весьма интересными молекулярными свойствами. Одно из таких наблюдаемых явлений носит название химический сдвиг. Химический сдвиг можно наблюдать только в ядрах атомов определенных типов.

Слово «сдвиг» применено, потому что измеряется сдвиг (смещение) сигнала по отношению к исследуемому значению, а «химическим» этот сдвиг назван потому, что он некоторым образом связан с химической средой исследуемых ядер.

```python
import numpy as np
import pymc3 as pm
import arviz as az
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

sns.set()
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
%load_ext watermark
%watermark --iversions
```

```python
data = np.loadtxt('data/chemical_shifts.csv')
az.plot_kde(data, rug=True)
plt.title('KDE. Chemical shifts')
plt.yticks([0], alpha=0)
plt.show()
```

График ЯОП (ядерной оценки плотности) для этого набора данных демонстрирует распределение, похожее на гауссово, за исключением двух точек данных, располагающихся далеко от среднего значения.

Предположим все же, опустив наличие этих двух точек, что гауссово распределение является правильным описанием исследуемых данных. Соответственно, можно создать следующую модель:

$$\mu \sim U(l, h);$$
$$\sigma \sim |N(0, \sigma_{\sigma})|;$$
$$y \sim N(\mu, \sigma).$$

```python
with pm.Model() as model_g:
    mu = pm.Uniform('mu', lower=40, upper=70)
    sigma = pm.HalfNormal('sigma', sd=10)
    y = pm.Normal('y', mu=mu, sd=sigma, observed=data)
    trace_g = pm.sample(draws=1000)
```

```python
az.plot_trace(trace_g)
plt.show()
```

```python
az.plot_joint(trace_g, kind='kde', fill_last=False)
plt.show()
```

```python
az.summary(trace_g)
```

После вычисления апостериорного распределения можно использовать эти результаты для имитационных данных и для проверки согласованности (связности) имитационных данных с наблюдаемыми данными.

```python
y_pred_g = pm.sample_posterior_predictive(trace_g, 100, model_g)
```

y_pred_g – это словарь, в котором ключи keys представлены именем наблюдаемой переменной в нашей модели, а значения values – это массив вида (samples, size) – (100, len(data)).

```python
data_ppc = az.from_pymc3(trace=trace_g, posterior_predictive=y_pred_g)
ax = az.plot_ppc(data_ppc, figsize=(6, 4), mean=False)
ax.legend(fontsize=15)
plt.show()
```

Полупрозрачные линии отображают неопределенность информации о статистически выведенном распределении прогнозируемых данных. Иногда при слишком малом количестве точек данных на подобном графике прогнозируемые кривые могут выглядеть как непричесанная копна.

Среднее значение имитационных данных слегка смещено вправо. Наблюдаемая дисперсия выглядит несколько большей для имитационных данных по сравнению с реальными данными. Это следствие нескольких смещенных точек.

Одним из недостатков, присущих модели `model_g`, является предварительное предположение о нормальном распределении, которое в сущности не выполняется из-за нескольких смещенных точек в данных. Можно считать, что эти точки имеют избыточный вес при определении параметров нормального распределения.

Чтобы поправить этот момент, можно

1. Использовать эмпирические правила по отсеву выбросов;
2. Изменить первоначальную модель.

Вариант изменения первоначальной модели кажется наиболее правильным с точки зрения концепций байесовского анализа.

Вариант обработки выбросов с использованием нормального распределения  - это замена на распределение Стьюдента (t-распределение). Параметры распределения: среднее значение, коэффициент масштаба (аналогичен
стандартному отклонению), число степеней свободы (иногда называемый параметром нормальности в интервале $[0, \infty]$). 

Чем больше степеней свободы - больше распределение приближается к нормальному, менее широкими становятся хвосты и наоборот. Именно хвосты t-распределения полезны для обработки возможных выбросов в таком байесовском анализе.

```python
plt.figure(figsize=(8, 4))
plt.title('t-distribution')
x_values = np.linspace(-10, 10, 500)
for df in [1, 2, 30]:
    distri = stats.t(df)
    x_pdf = distri.pdf(x_values)
    plt.plot(x_values, x_pdf, label=fr'$\nu = {df}$', lw=3)
    
x_pdf = stats.norm.pdf(x_values)
plt.plot(x_values, x_pdf, 'k--', label=r'$\nu = \infty$')  
plt.xlabel('x')
plt.yticks([])
plt.legend()
plt.xlim(-5, 5)
```

Новая модель:

$$\mu \sim U(l, h);$$
$$\sigma \sim |N(0, \sigma_{\sigma})|;$$
$$\nu \sim Exp(\lambda);$$
$$y \sim \tau(\mu, \sigma, \nu).$$

```python
with pm.Model() as model_t:
    mu = pm.Uniform('mu', 40, 75)
    sigma = pm.HalfNormal('sigma', sd=10)
    nu = pm.Exponential('nu', 1/30)
    y = pm.StudentT('y', mu=mu, sd=sigma, nu=nu, observed=data)
    trace_t = pm.sample(1000)
```

```python
az.plot_trace(trace_t)
plt.show()
```

```python
az.summary(trace_t)
```

Проверка прогнозируемого апостериорного распределения

```python
y_ppc_t = pm.sample_posterior_predictive(
trace_t, 100, model_t, random_seed=123)
y_pred_t = az.from_pymc3(trace=trace_t, posterior_predictive=y_ppc_t)
ax = az.plot_ppc(y_pred_t, figsize=(6, 4), mean=False)
plt.legend(fontsize=15)
plt.xlim(40, 70)
```
