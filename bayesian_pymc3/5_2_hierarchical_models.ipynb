Будем измерять только определенные типы атомных ядер. Протеины состоят из цепочек размером в 20 элементов, известных как остатки или радикалы аминокислот. Каждая аминокислота может появляться в такой цепочке ноль или более раз, поэтому длина цепочек варьируется от нескольких до сотен и тысяч остатков аминокислот.

Каждая из 20 аминокислот имеет различные химические свойства, определяющие биологические характеристики протеина, некоторые могут обладать суммарными электрическими зарядами, другие абсолютно нейтральны.


Здесь основной факт состоит в том, что аминокислоты похожи, но не одинаковы, следовательно, вполне разумным или даже естественным будет обоснование любого вывода, относящегося к химическому сдвигу, в терминах 20 групп в соответствии с определением типов аминокислот.


Реальные эксперименты слишком сложны, и всегда существует вероятность того, что не удастся получить полную информацию о химическом сдвиге.


Данные:

* ID: идентификатор протеина;
* aa: названия аминокислот, буквенный код;
* theo: теоретически вычисленные значений химического сдвига;
* exp:  результаты экспериментальных замеров;

Одна из целей этого примера – сравнение различий, чтобы узнать, насколько точно теоретические вычисления соответствуют экспериментальным измерениям.

```python
df_cs = pd.read_csv('data/chemical_shifts_theo_exp.csv')
```

```python
df_cs.head()
```

Для оценки различий в химических сдвигах, подсчитанных теоретически и после реальных замеров, используем как иерархическую, так и не ирерхическую вероятностную модель с целью их сопоставления.

```python
diff = df_cs.theo.values - df_cs.exp.values
idx = pd.Categorical(df_cs['aa']).codes
groups = len(np.unique(idx))
```

```python
idx
```

```python
groups
```

Модель сравнения групп:

```python
with pm.Model() as cg_model:
    mu = pm.Normal('mu', mu=0, sd=10, shape=groups)
    sigma = pm.HalfNormal('sigma', sd=10, shape=groups)
    
    y = pm.Normal('y', mu=mu[idx], sd=sigma[idx], observed=diff)
    
    trace_cg = pm.sample(1000)
```

```python
az.summary(trace_cg)
```

```python
az.plot_trace(trace_cg, figsize=(12,5))
plt.show()
```

Иерархическая модель: добавляем два гипер-априорных распределения:

* гипер-априорное распределение для среднего $\mu$
* гипер-априорное распределение для стандартного отклонения $\mu$

В качестве показательного примера выбираем не максимально возможно сложную иерархическую модель, т.е. $\sigma$ оставляем без гипер-априорных распределений

```python
with pm.Model() as h_model:
    
    # Гипер-априорные распределения
    mu_mu = pm.Normal('mu_mu', mu=0, sd=10)
    sigma_mu = pm.HalfNormal('sigma_mu', 10)
    
    # Априорные распределения
    mu = pm.Normal('mu', mu=mu_mu, sd=sigma_mu, shape=groups)
    sigma = pm.HalfNormal('sigma', sd=10, shape=groups)
    
    y = pm.Normal('y', mu=mu[idx], sd=sigma[idx], observed=diff)
    
    trace_h = pm.sample(1000)
```

```python
az.summary(trace_h)
```

```python
az.plot_trace(trace_h, figsize=(12,5))
plt.show()
```

Также напрямую сопоставим результаты двух моделей с помощью arviz:

```python
axes = az.plot_forest([trace_cg, trace_h],
                      model_names=['non_hierarch', 'hierarch'],
                      var_names='mu', combined=False, 
                      colors='cycle',
                      credible_interval=0.95,
                     )

y_lims = axes[0].get_ylim()
axes[0].vlines(trace_h['mu_mu'].mean(), *y_lims)

plt.show()
```

Таким образом, с помощью arviz.plot_forest получили график 40 оцененных средних - по 20 сдвигов теоретических значений от экспериментальных для 2х разных моделей, с учетом 95-процентного вероятностного интервала наименьшей длины (credible interval) и интерквартильного размаха в виде более жирной центральной части каждой линии.

Вертикальной линией отметили общее среднее для иерархической модели.

Самым важным на этом графике является то, что оценки, полученные от иерархической модели, подтягиваются к частично объединенному среднему значению, или, что одно и то же, редуцируются с учетом необъединенных оценок.

Эффект более заметен для групп, расположенных дальше от среднего значения (например, 13),неопределенность равна или меньше, чем неопределенность в неиерархической модели.

Таким образом, мы получаем промежуточное состояние между наличием одной группы, объединяющей все химические сдвиги, и наличием 20 отдельных групп, по одной для каждой аминокислоты.
